```python
data = pd.read_json('./data/challenge_3.json')
#Splitting the city names in the column (cities) into lists for easy of handling
data['cities'] = data['cities'].apply(lambda r:[e.strip() for e in r[0].split(',')])

#Get the unique cities
unique_cities = set(data.cities.sum())
```

```python
#The lines below will return the city 
#name and the one most often was searched with

freq_two_cities = defaultdict(lambda: defaultdict(int))

for s in data['cities'][:]:
    #Skipping sessions has only one city searched
    if len(s) == 1:
        continue
    #Some sessions have more than two cities searched
    candidates = combinations(s, 2)
    for c in candidates:
         freq_two_cities[c[0]][c[1]] +=1

#Return the city with the max occurance for each city
for city, other_cities in freq_two_cities.items():
    other_city = max(other_cities, key=other_cities.get)
    print(city, ' ------> ',other_city, ' -------> ', other_cities[other_city])
```

<h3 style="color:red"> Travel sites are browsed by two kinds of users. Users who are actually planning a  trip and users who just dream about a vacation. The first ones have obviously a much higher purchasing intent. Users planning a trip often search for cities close to each other, while users who search for cities far away from each other are often just dreaming about a vacation. Based on this idea, can you come up with an algorithm that clusters sessions into two groups: high intent and low intent.
</h3>

<h3 style="color:blue">One possible solution is to spatially cluster the cities that were searched. 
Each cluster will be given a label. Then, go through each session to find the cluster label of each searched city. If all the searched cities belong to one cluster lable, then that session is of high intent, otherwise it is a low intent session. If there is only one searched city then 'unknown' intent is returned to that session. 
</h3>

```python
#Step 1: Get the latitude and longitude of each city from google map API
cities_coordinates = []
for city in unique_cities:
    location = geolocator.geocode(city, language='en', timeout=30)
    print(city, ' latitude:', location.latitude,' longitude:',location.longitude)
    cities_coordinates.append([city, location.latitude, location.longitude])
```

```python
#Just intermediate step
cities_df = pd.DataFrame(cities_coordinates, columns=['city', 'latitude', 'longitude'])
coords = cities_df.as_matrix(columns=['latitude','longitude'])
```

```python
#Step 2:Using DBSCAN from sklearn to do clustering, note epsilon will decide the size of the cluster 
kms_per_radian = 6371.0088
epsilon = 200 / kms_per_radian
db = DBSCAN(eps=epsilon, min_samples=1, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
cluster_labels = db.labels_
num_clusters = len(set(cluster_labels))
clusters = pd.Series([coords[cluster_labels == n] for n in range(num_clusters)])
print('Number of clusters: {}'.format(num_clusters))
```

```python
#Attach the cluster label for each city
cities_df['cluster_label'] = cluster_labels
```

```python
#Step 3:Going through the sessions and check if the searched cities belong to the same cluster
for index, row in data.iterrows():
    searched_cities = row['cities']
    if len(searched_cities) == 1:
        print(row['session_id'][0], ' ---> ','unknown:', ' ---> ',row['cities'])
        continue
    #Get the label of the searched cities
    searched_cities_labels = [cities_df[cities_df['city']==c]['cluster_label'].values[0] for c in searched_cities]
    if len(set(searched_cities_labels)) == 1:
        print(row['session_id'][0], ' ---> ', 'high intent:', ' ---> ', row['cities'])
    else:
        print(row['session_id'][0], ' ---> ','low intent:', ' ---> ', row['cities'])
   
```
